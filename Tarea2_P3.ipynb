```python
x_1 = Symbol('x_1')
x_2 = Symbol('x_2')

#Tarea 1
#ùêπ(ùë•_1,ùë•_2 ) = (ùë•_1‚àí1)^2+(ùë•_2‚àí1)^4; (ùë•_1+1)^2+(ùë•_2+1)^2] 
#con ùí≥=[‚àí3, 3]√ó[‚àí3, 3]
#(1-1, 1-2, 3, 4 y 5)

gamma = 0.7 
f1 = 0.5 * (( ( 1 + (x_1 + x_2)**2 )**0.5 ) + ( ( 1 + (x_1 - x_2)**2 )**0.5 ) + x_1 - x_2) + gamma * exp(-(x_1-x_2)**2)
f2 = 0.5 * (( ( 1 + (x_1 + x_2)**2 )**0.5 ) + ( ( 1 + (x_1 - x_2)**2 )**0.5 ) - x_1 + x_2) + gamma * exp(-(x_1-x_2)**2)
```

```python
#graficamente
def fg1(x, y):
  gamma = 0.7 
  return 0.5 * (( ( 1 + (x + y)**2 )**0.5 ) + ( ( 1 + (x - y)**2 )**0.5 ) + x - y) + gamma * np.exp(-(x-y)**2)
  #return 0.5 * ( sqrt( (1+(x+y)**2) ) + sqrt((1+(x-y)**2) ) + x - y ) + gamma * exp(-(x-y)**2)
      
def fg2(x, y):
  gamma = 0.7 
  return 0.5 * (( ( 1 + (x + y)**2 )**0.5 ) + ( ( 1 + (x - y)**2 )**0.5 ) - x + y) + gamma * np.exp(-(x-y)**2)
  #return   0.5 * ( sqrt( (1+(x+y)**2) ) + sqrt((1+(x-y)**2) ) - x + y ) + gamma * exp(-(x-y)**2)


xg = np.linspace(-2, 2, 50)
yg = np.linspace(-2, 2, 50)

Xg, Yg = np.meshgrid(xg, yg)
Zg1 = fg1(Xg, Yg)
Zg2 = fg2(Xg, Yg)


fig = plt.figure()
ax = plt.axes(projection='3d')

ax.view_init(50, 70)

ax.contour3D(Xg, Yg, Zg1, 50, cmap='Blues')
ax.contour3D(Xg, Yg, Zg2, 50, cmap='Reds')

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z');
```

```python
def Gradient(fx, X):
  return np.array( [ [fx.diff(x_1).subs( [(x_1, float(X[0])), (x_2, float(X[1]))]  )], [fx.diff(x_2).subs( [(x_1, float(X[0])), (x_2, float(X[1]))]  ) ] ] )  

def Direction(fx, X):  
  dy = -1 * Gradient(fx, X)
  return dy

def Step_Size(fx, X, alpha, ro, c):
  counter = 0
  origen_0 = fx.subs([(x_1, float(X[0])), (x_2, float(X[1]))]).doit()
  f_k = origen_0
  
  pk = Direction(fx, X)
  
  g = Gradient(fx, X)

  while True:
    dato1 = X  + (alpha * pk)
    destino = fx.subs([(x_1, float(dato1[0])), (x_2, float(dato1[1]))]).doit()

    dato3 = (c * alpha) * np.dot(g.T, pk)
    penal_0 = dato3[0,0]

    f_n = round(float(destino), 6)
    penalty = round(float(penal_0), 6)

    if (f_n <= f_k + penalty) or (counter>100):
      return alpha

    alpha = round(ro * alpha, 6)
    counter = counter + 1
```

```python
def BasicFiltering(Poblacion_r):
  A = []

  for p in range(len(Poblacion_r)):
    agrega = 1
    for a in range(len(Poblacion_r)):
      if a!=p:
        #print("verifica dominancia")
        x_p = Poblacion_r[p][0]
        y_p = Poblacion_r[p][1]

        x_a = Poblacion_r[a][0]
        y_a = Poblacion_r[a][1]

        f_1_p = f1.subs([ (x_1, x_p), (x_2, y_p) ] )
        f_2_p = f2.subs([ (x_1, x_p), (x_2, y_p) ] )

        f_1_a = f1.subs([ (x_1, x_a), (x_2, y_a) ] )
        f_2_a = f2.subs([ (x_1, x_a), (x_2, y_a) ] )

        if (f_1_a<f_1_p) & (f_2_a<f_2_p):
          agrega=0

    if agrega==1:
      A.append(Poblacion_r[p])

  return A
```

```python
#filtrado de soluciones
#crear 100,000 soluciones aleatorias

for k in range(1000):
  graph1=[]
  graph2=[]

  num_sols = 10
  Poblacion = []

  for k in range(num_sols):
    x1_rand = round(random.uniform(-3, 3), 2)
    x2_rand = round(random.uniform(-3, 3), 2)

    Poblacion.append([x1_rand, x2_rand])

  A = BasicFiltering(Poblacion)  

  f_1 = f1.subs([ (x_1, A[0][0]), (x_2, A[0][1]) ] )
  f_2 = f2.subs([ (x_1, A[0][0]), (x_2, A[0][1]) ] )

  graph1.append(f_1)
  graph2.append(f_2)

  plt.scatter(graph1, graph2)


  
ax = pylab.subplot(111)

ax.set_xlabel('f1')
ax.set_ylabel('f2')



plt.show()
```
